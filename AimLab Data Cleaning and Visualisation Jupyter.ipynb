```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df = pd.read_csv(r'C:\Users\david\PycharmProjects\NewPc\AimLab\AimlabTaskData_2023-02-09.01.30.45\GridshotUltimate.csv')

cols =  df.columns

new_df =  df.drop(["accB0", "accB1", "accB2", "accB3", "accB4", "accB5", "accB6", "accB7", 'rtB0', 'rtB1', 'rtB2',
                   'rtB3', 'rtB4', 'rtB5', 'rtB6', 'rtB7',  "version", "map", "mode", "weaponName"], axis = 1)

yo = new_df.rename(columns={"targetsTotal": "Total Shots",
                            "shotsTotal": "Hits Per Sec",
                            "killsPerSec": "Total Hits",
                           "accTotal": "Accuracy Rate",
                           "createDate": "Task",
                           "rtTotal": "Reaction Time(ms)",
                           "taskName": "Task Name"})

yoV2 = yo.drop(["Task Name", "killTotal"], axis=1)



yoV2.iloc[95:, [1, 2]] = yoV2.iloc[95:, [2, 1]]

print(yoV2)
```

```python
plt.rcParams['figure.figsize'] = [24, 5]

yoV2 = yoV2[yoV2["Total Hits"] >= 100]

visual_total_hits = yoV2.plot.line(y = "Total Hits")

visual_total_hits.set_xlim(visual_total_hits.get_xlim()[::-1])

plt.show()
```

```python
plt.rcParams['figure.figsize'] = [24, 5]

visual_hits_per_sec = yoV2.plot.line(y = "Hits Per Sec")

visual_hits_per_sec.set_xlim(visual_hits_per_sec.get_xlim()[::-1])

plt.show()
```

```python
plt.rcParams['figure.figsize'] = [24, 5]

yoV3 = yoV2[yoV2['Accuracy Rate'] <= 101]

visual_accuracy_rate = yoV3.plot.line(y = "Accuracy Rate")

visual_accuracy_rate.set_xlim(visual_accuracy_rate.get_xlim()[::-1])

plt.show()
```

```python
plt.rcParams['figure.figsize'] = [24, 5]

visual_total_shots = yoV2.plot.line(y = "Total Shots")

visual_total_shots.set_xlim(visual_total_shots.get_xlim()[::-1])

plt.show()
```

```python
plt.rcParams['figure.figsize'] = [24, 5]

visual_reaction_time = yoV2.plot.line(y = "Reaction Time(ms)")

visual_total_shots.set_xlim(visual_total_shots.get_xlim()[::-1])

plt.show()
```
